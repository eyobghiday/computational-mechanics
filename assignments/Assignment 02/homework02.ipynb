$Part$ $A$

```python
def vFunc(r):
 return sp.sin(r) + r ** 2 + 1
r=sp.Symbol('r')
#to find the expression of first, second and third differential
print(sp.diff(vFunc(r),r))
print(sp.diff(vFunc(r),r,r))
print(sp.diff(vFunc(r),r,r,r))
```

$Part$ $D$

```python
r=np.arange(-np.pi,np.pi,0.05)
w=2*r + np.cos(r)
plt.xlabel("r in radians")
plt.ylim(0,7,0.5)
plt.plot(r,w,'b')
plt.legend(['dv(r)/dr'])
plt.show()
```

The Value of the "r" for which function "w" is 0 is approximately -0.46. This is equal with the result in Part C. Because the the dervateive of of a function solved at zero would give us the value at which the original function is minumum. In other words the Minimum values of a function can be found using by taking it's first dervative and equating it to zero.

```python
#Taylor Series approximation 
x=np.arange(0,2*np.pi,0.05)
x0=0
yacc=np.sin(x)+np.cos(x)
y2=np.sin(x0)+np.cos(x0)+(x-x0)*(np.cos(x0)-np.sin(x0))
y3=y2+((x-x0)**2)*(-np.cos(x0)-np.sin(x0))/2
y4=y3+((x-x0)**3)*(-np.cos(x0)+np.sin(x0))/6
y5=y4+((x-x0)**4)*(np.cos(x0)+np.sin(x0))/24
y6=y5+((x-x0)**5)*(np.cos(x0)-np.sin(x0))/120
plt.plot(x,yacc,x,y2,x,y3,x,y4,x,y5,x,y6)
plt.show()
```

```python
#Let's define the function F(x)=x^3(cos(x))
def func(x):
    y = (x**3)*np.cos(x)
    return y

x=fsolve(func,-18)
print(x)
x=fsolve(func,-15)
print(x)
x=fsolve(func,-12)
print(x)
x=fsolve(func,-8)
print(x)
x=fsolve(func,-4)
print(x)
x=fsolve(func,-2)
print(x)
x=fsolve(func,1)
print(x)
x=fsolve(func,2)
print(x)
x=fsolve(func,5)
print(x)
x=fsolve(func,7)
print(x)
x=fsolve(func,11)
print(x)
x=fsolve(func,14)
print(x)
x=fsolve(func,16)
print(x)
print("We see that it has found approximated the 13 roots of the function")
```

$PROBLEM$ $5$

```python
def f(z):
    x=z[0]
    y=z[1]
    f=np.zeros(2)
    f[0]=y+2*x+4
    f[1]=y-x**2+4
    return f
zGuess=np.array([-1,1])
z=fsolve(f,zGuess)
print ("First set of Solutions")
print(z,"\n")


zGuess=np.array([0,4])
print ("Second set of Solutions")
z=fsolve(f,zGuess)
print(z)
```
