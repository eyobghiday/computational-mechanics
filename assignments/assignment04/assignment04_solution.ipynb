$Part$ $B$

```python
a0=4
a1=1.5
a2=0.3
x=np.arange(-20,20,1).reshape((-1,1))
n=np.array(np.random.uniform(-10,10,40))
# n=np.array(10*(np.random.rand()-0.5)).T

y=a0+a1*x+a2*(x**2)+n
transformer = PolynomialFeatures(degree=2, include_bias=False)
x_=transformer.fit_transform(x)
# print(x)
# print(x_)
model = LinearRegression().fit(x_, y)
r_sq = model.score(x_, y)

x_new = np.arange(-20,21).reshape((-1, 1))
x_new_=transformer.fit_transform(x_new)
y_pred = model.predict(x_new_)
plt.plot(x,y,'ro',x_new,y_pred)
plt.show()
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('coefficients:', model.coef_)
```

Comparing the predicted a0,a1,a2 to the orignal are the same wtih the excpetion of a0, and that is de to the random noise data introduced through the value of n.

```python
from scipy.io import loadmat
data = loadmat('data.mat')
x=data['x'].reshape((-1,1))
y=data['y'].reshape((-1,1))
transformer=PolynomialFeatures(degree=3, include_bias=False)
x_=transformer.fit_transform(x)
model=LinearRegression().fit(x_,y)
r_sq=model.score(x_,y)
x_new = np.arange(-21,21).reshape((-1, 1))
x_new_=transformer.fit_transform(x_new)
y_pred=model.predict(x_new_)
plt.plot(x,y,'ro',x_new,y_pred)
plt.show()
```

Thus we can see that a polynomial of degree 3 is best fit for the data shown above.
